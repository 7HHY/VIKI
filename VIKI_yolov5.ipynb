## WandB 시각화
- 모델 훈련을 시작하면 다음과 같은 형식이 나온다.
```
wandb: (1) Create a W&B account
wandb: (2) Use an existing W&B account
wandb: (3) Don't visualize my results
wandb: Enter your choice: (30 second timeout) 
```
- 1번은 새로운 wandb 계정을 만드는 옵션
- 2번은 wandb 계정이 있다면 로그인하고 wandb를 연결하는 옵션
- 3은 시각화를 건너뛰는 옵션이다.
- 계정을 만들고 https://wandb.ai/authorize 에서 API키를 얻어온 후 입력해주면 wandb와 연동이 시작된다. (실시간으로 과정 모니터링 시각화 가능)
